```python
import networkx as nx
import pandas as pd
import pickle
```

```python
import pickle

with open('collaborations.pkl', 'rb') as f:
    colabs = pickle.load(f)
```

```python
artists = pd.read_csv('../spotify/spotify_artists.csv')
filtered_artists = artists[artists['popularity'] >= 65]
filtered_artists.reset_index(drop=True, inplace=True)
filtered_artists['collaborations'] = colabs
filtered_artists
```

```python
name_to_id = {}
for i in range(len(filtered_artists)):
    name_to_id[filtered_artists['name'][i]] = filtered_artists['id'][i]
```

```python
filtered_artists['name'].value_counts()
```

```python
c = ['UZI', 'Eve', 'Chris Brown']
mudar = []
for i in range(len(filtered_artists)):
    for x in c:
        if x in list(filtered_artists['collaborations'][i]):
            mudar.append(i)
```

```python
print(mudar)
```

```python
with open('mudar.pkl', 'wb') as arquivo:
    pickle.dump(mudar, arquivo)
```

```python
with open('nova_colab.pkl', 'rb') as arquivo:
    itera = pickle.load(arquivo)
```

```python
pos = 0
colab_ids = []
artists = list(filtered_artists['name'])
artists_ids = list(filtered_artists['id'])
for i in range(len(filtered_artists)):
    if i in mudar:
        ids = itera[pos]
        pos += 1
        string = ''
        for i in ids:
            if i in artists_ids:
                string += f'{i}, '
    else:
        nomes = filtered_artists['collaborations'][i]
        string = ''
        for n in nomes:
            if n in artists:
                id = name_to_id[n]
                string += f'{id}, '
    colab_ids.append(string)
```

```python
colab_ids
```

```python
filtered_artists['collaborations_ids'] = colab_ids
filtered_artists
```

```python
filtered_artists.to_csv('collaborations.csv')
```
