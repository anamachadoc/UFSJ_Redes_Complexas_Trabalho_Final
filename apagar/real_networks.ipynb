### **Rede de pessoas**

```python
network_real_people = nx.read_edgelist('../networks/people/people_edges_list.csv', delimiter = ',', nodetype = int, create_using = nx.Graph()) 
```

#### **Informações básicas**

```python
if not os.path.exists(f'../data/real/people'): os.makedirs(f'../data/real/people')
dir = '../data/real/people/dados_real_people.csv'
```

```python
dados = pd.DataFrame()
nodes = sorted(network_real_people.nodes)
dados['nodes'] = nodes
dados.to_csv(dir, index = False)
```

```python
printInformations(network_real_people)
```

#### **Distribuição de graus**

```python
frequencies_of_degrees = nx.degree_histogram(network_real_people)
createGraphic(data = frequencies_of_degrees, xLabel = 'Grau (K)', yLabel = 'Frequência', title = 'Frequência dos Graus', limit = 5, lines = False, name_network = 'people')

probability_of_degrees = [degree/(nx.number_of_nodes(network_real_people)) for degree in frequencies_of_degrees]
createGraphic(data = probability_of_degrees, xLabel = 'Grau (K)', yLabel = 'P(k)', title = 'Probabilidade dos Graus', limit = 0.01, lines = False,  name_network = 'people')

accumulated_probability_of_degress = [sum(probability_of_degrees[index:]) for index in range(len(probability_of_degrees))]
createGraphic(data = accumulated_probability_of_degress, xLabel = 'Grau (k)', yLabel = 'P(>=K)', title = 'Função de Distribuição Cumulativa Complementar', limit = 0.05, lines = True,  name_network = 'people')
```
