```python
client_id = '4044bebb349746dab35811484f587d7a'
client_secret='e2a8b70bfe91496da622a223a1284ff6'
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id, 
                                                                        client_secret=client_secret))
```

```python
df = pd.read_csv('spotify_artists.csv')
df
```

```python
df = df.dropna(subset=['genres'])
df.reset_index(drop=True, inplace=True)
df
```

```python
with open('artist_related_artists.csv', 'a+') as file:
    file.seek(0)
    content = file.readlines()
    for i in tqdm(range(len(content), len(df))):
        artist_related_artists = sp.artist_related_artists(df['id'][i])
        artist_related_artists_str = ', '.join([artist['id'] for artist in artist_related_artists['artists']])
        file.write(f"{df['id'][i]}\t{(artist_related_artists_str)}")
        file.write('\n')
```
